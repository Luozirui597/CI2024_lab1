```python
import numpy as np
'''I got error message on vscode:
A module that was compiled using NumPy 1.x cannot be run inNumPy 2.1.1 as it may crash. 
To support both 1.x and 2.xversions of NumPy, modules must be compiled with NumPy 2.0.
Some module may need to rebuild instead e.g. with 'pybind11>=2.12'.
So,I changed to plotly with friend's advice'''
import plotly.graph_objects as go
from tqdm import tqdm
```

```python
#set size 
UNIVERSE_SIZE = 1000
NUM_SETS = 100
DENSITY = 0.2

SETS = np.random.random((NUM_SETS, UNIVERSE_SIZE)) < DENSITY
```

```python
# DON'T EDIT THESE LINES!

SETS = np.random.random((NUM_SETS, UNIVERSE_SIZE)) < DENSITY
for s in range(UNIVERSE_SIZE):
    if not np.any(SETS[:, s]):
        SETS[np.random.randint(NUM_SETS), s] = True
COSTS = np.pow(SETS.sum(axis=1), 1.1)
```

```python
def valid(solution):
    """Checks wether solution is valid (ie. covers all universe)"""
    return np.all(np.logical_or.reduce(SETS[solution]))


def cost(solution):
    """Returns the cost of a solution (to be minimized)"""
    return COSTS[solution].sum()
```

```python
def tweak(solution):
  new_solution = solution.copy()
  flip_set = np.random.randint(0,NUM_SETS)
  new_solution[flip_set] = not new_solution[flip_set]
  return new_solution
```

```python
def hill_climbing(steps):
  current_solution = np.random.choice([True, False], size=NUM_SETS, p=[0.5, 0.5])
    
  history = []
  for _ in tqdm(range(steps)):
    new_solution = tweak(current_solution)

    if valid(new_solution) and cost(new_solution) < cost(current_solution):
        current_solution = new_solution
    history.append(cost(current_solution))

    print(f"Current cost: {cost(current_solution)}")

  return current_solution,history
```

```python
best_solution,history = hill_climbing(1000)
```

```python
#draw the image
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history))), y=history, mode='lines+markers', name='Cost History'))
fig.update_layout(title='Hill Climbing Cost History', xaxis_title='Steps', yaxis_title='Cost')
fig.show()

#final
print(f"Best solution: {cost(best_solution)}")
```
