```python
# DON'T EDIT THESE LINES!

SETS = np.random.random((NUM_SETS, UNIVERSE_SIZE)) < DENSITY
for s in range(UNIVERSE_SIZE):
    if not np.any(SETS[:, s]):
        SETS[np.random.randint(NUM_SETS), s] = True
COSTS = np.pow(SETS.sum(axis=1), 1.1)
```

Try Simulated Annealing first(Because it looks like powerful :)

```python
def valid(solution):
    """Checks wether solution is valid (ie. covers all universe)"""
    return np.all(np.logical_or.reduce(SETS[solution]))


def cost(solution):
    """Returns the cost of a solution (to be minimized)"""
    return COSTS[solution].sum()
```

```python
def Simulate_Annealing(initial_solution,steps,initial_temp,cool_rate):

  current_solution = initial_solution
  best_solution = initial_solution
  current_temp = initial_temp

  for step in range(steps):
    # randomly choose a set
    new_solution = current_solution.copy()
    # like I learned from lecture,try to flip those sets for add more solutions
    flip_set = np.random.randint(0, NUM_SETS)
    new_solution[flip_set] = not new_solution[flip_set]

    # if the new solution works,then start to record and replace
    if valid(new_solution):
      new_cost = cost(new_solution)
      current_cost = cost(current_solution)
      # decide whether to accept new solution or not,and try to avoid local optimal solution
      if new_cost < current_cost or np.random.rand() < np.exp((current_cost - new_cost)/current_temp):
        current_solution = new_solution
        if new_cost <cost(best_solution):
          best_solution = new_solution

    # then be cool~
    current_temp *= cool_rate

    ic(step,cost(best_solution))
    #print(f"Step: {step}, Cost: {cost(best_solution)}")
  return best_solution
```

```python
#initial solution
initial_solution = np.random.rand(NUM_SETS) < 0.4
```

```python
#start
best_solution = Simulate_Annealing(
  initial_solution,
  steps = 500,
  initial_temp = 100,
  cool_rate = 0.78
)

ic(valid(best_solution),cost(best_solution))
#print(f"valid: {best_solution}, Cost: {cost(best_solution)}")
```
